# 🌍 Análisis de Emisiones de CO₂ (1750–2021)
### Dataset: Global Carbon Project – GCB2022v27_MtCO2_flat.csv

El presente notebook analiza la evolución histórica de las emisiones globales de dióxido de carbono (CO₂) por tipo de fuente y por país, a partir de los datos del **Global Carbon Project**.

El **Global Carbon Project (GCP)** es una iniciativa científica internacional dedicada a **cuantificar las emisiones y sumideros de carbono** a nivel global, con el fin de **apoyar políticas climáticas basadas en evidencia**.

🔗 **Fuente oficial del dataset:** [Global Carbon Project – CO₂ Emissions Dataset](https://www.globalcarbonproject.org/carbonbudget)

El objetivo es ofrecer una visión integral sobre la evolución de las emisiones globales de CO₂, sus principales fuentes y las diferencias entre países a lo largo del tiempo.

---

## 📑 Descripción del dataset

El dataset contiene las emisiones anuales de CO₂ (en **millones de toneladas – MtCO₂**) desagregadas por país y por fuente de emisión.  

Las principales columnas son:

- **Country** → País emisor.  
- **Year** → Año del registro.  
- **Total** → Emisiones totales de CO₂ del país en ese año.  
- **Coal** → Emisiones derivadas del uso de carbón.  
- **Oil** → Emisiones derivadas del uso de petróleo.  
- **Gas** → Emisiones derivadas del uso de gas natural.  
- **Cement** → Emisiones asociadas a la producción de cemento.  
- **Flaring** → Emisiones por quema de gases residuales en la extracción de petróleo y gas.

> ℹ️ Los datos cubren más de 150 países y abarcan un periodo extenso (1750–2021).  
> En este notebook se trabaja principalmente desde **1925 en adelante**, para centrarse en la era industrial moderna y contar con registros más consistentes.

## 🧹 Tratamiento de Datos

```python
# 1. Recalcular 'Total' cuando está nulo, sumando fuentes disponibles
df["Total"] = df["Total"].fillna(
    df[["Coal", "Oil", "Gas", "Cement", "Flaring", "Other"]].sum(axis=1)
)

# 2. Imputar columnas de fuentes con la media por país (solo si hay valores disponibles)
for col in ["Coal", "Oil", "Gas", "Cement", "Flaring"]:
    df[col] = df.groupby("Country")[col].transform(lambda x: x.fillna(x.mean()))

# 3. Eliminar columnas con más del 90% de nulos (poca utilidad para análisis general)
df.drop(columns=["Other", "Per Capita"], inplace=True)
```

```python
# 4. Confirmar que la limpieza se aplicó correctamente
print("Nulos restantes por columna:")
print(df.isnull().sum())
```

```python
# Ver países con datos faltantes en esas columnas
nulos_df = df[df[["Coal", "Gas", "Cement", "Flaring"]].isnull().any(axis=1)]
print(nulos_df["Country"].value_counts())
```

```python
# Eliminar filas con cualquier nulo restante en columnas clave
df.dropna(subset=["Coal", "Gas", "Cement", "Flaring"], inplace=True)

# Verificación final
print("Nulos restantes por columna:")
print(df.isnull().sum())
```

```python
# Eliminar columna con muchos nulos que no es esencial por ahora
df.drop(columns=["ISO 3166-1 alpha-3"], inplace=True)

# Verificar nulos restantes en el DataFrame
print("Verificación final - Nulos por columna:")
print(df.isnull().sum())
print(f"\nTamaño final del dataset: {df.shape[0]} filas × {df.shape[1]} columnas")
```

## 🌎 Visualizaciones Globales (1750–2021)

**Emisiones globales de CO₂ a través del tiempo (1750 - 2021)**

```python
#Evolución global de emisiones de CO₂ a lo largo del tiempo
import plotly.express as px

global_by_year = df.groupby("Year")["Total"].sum().reset_index()

fig = px.line(global_by_year, x="Year", y="Total", title="🌍 Emisiones globales de CO₂ por año", labels={"Total": "Emisiones (MtCO₂)"})
fig.show()
```

**Interpretación: Emisiones globales de CO₂ por año (1750–2021)**

- **1750–1850:** Las emisiones se mantienen en niveles muy bajos y casi imperceptibles, lo que refleja una economía basada en procesos productivos pre-industriales y con registros escasos en ese periodo.

- **1850–1900:** Se inicia un crecimiento visible de emisiones, asociado a la Revolución Industrial, la expansión del uso del carbón y el aumento de la producción fabril.

- **1900–1950:** Se observa un crecimiento sostenido con algunas caídas puntuales en 1932 (Gran Depresión) y 1945 (Segunda Guerra Mundial), pero con tendencia clara al alza durante la primera mitad del siglo XX.

- **1950 en adelante:** Las emisiones aumentan de manera casi exponencial, impulsadas por la industrialización global, el auge del petróleo y la expansión del consumo energético. Solo se aprecian leves descensos en años críticos como 1983 (crisis económica) y 2020 (COVID-19).

➡️ En conjunto, este gráfico muestra la aceleración histórica del cambio climático: el crecimiento de las emisiones de CO₂ está directamente vinculado a la industrialización y al modelo de desarrollo económico global.

**Top 10 países con mayores emisiones totales**

```python
emisiones_fuentes = df.groupby("Year")[fuentes].sum().reset_index()

fig = px.line(emisiones_fuentes, x="Year", y=fuentes, title="📊 Evolución de emisiones por tipo de fuente")
fig.show()
```

### Análisis del gráfico

- **1750–1855:**  
El gráfico muestra que, en este periodo inicial, las emisiones asociadas al **petróleo (Oil)** aparecen levemente por encima del **carbón (Coal)** y del **gas (Gas)**. Sin embargo, este comportamiento refleja más una **limitación en los registros históricos** que una realidad productiva, ya que el uso masivo de petróleo no se consolidó sino hasta mediados del siglo XIX.  
Históricamente, el **carbón** fue la primera fuente energética con impacto global en las emisiones, especialmente a partir de la Revolución Industrial. En este tramo, las emisiones totales permanecen casi imperceptibles, reflejando una economía preindustrial y una muy baja intensidad energética.

- **1855–1950:**  
El **carbón (Coal)** adquiere un rol predominante, con un aumento sostenido y marcado en su pendiente, consolidándose como la principal fuente de emisiones durante este periodo.  
En contraste, las emisiones de **petróleo (Oil)** crecen a un ritmo más moderado, al igual que las de **gas (Gas)**, que mantienen una tendencia estable.  
Las fuentes **Cement** y **Flaring** continúan con niveles muy bajos y prácticamente constantes.

- **1950–2021:**  
A partir de mediados del siglo XX, se observa una **aceleración significativa** de las emisiones provenientes del **carbón**, cuya curva adopta una forma casi exponencial, seguida por un incremento notable del **petróleo (Oil)**.  
Sin embargo, desde el año 2000, el petróleo muestra una **leve desaceleración**, mientras el carbón se consolida como la fuente más contaminante a nivel global.   El **gas (Gas)** mantiene un crecimiento sostenido, aunque sin alcanzar los niveles de las dos fuentes principales. Finalmente, **Cement** y **Flaring** registran un aumento moderado desde los años 80, pero continúan representando una fracción menor dentro del total global de emisiones.

**Distribución de emisiones por tipo de fuente (1925–2021)**

```python
# Distribución de emisiones por tipo de fuente desde 1925
import plotly.express as px

# Lista de fuentes
fuentes = ["Coal", "Oil", "Gas", "Cement", "Flaring"]

# Agrupamos por año y sumamos emisiones por fuente
df_fuentes = df_1925.groupby("Year")[fuentes].sum().reset_index()

# Crear gráfico de áreas apiladas
fig = px.area(df_fuentes,
              x="Year",
              y=fuentes,
              title="🧮 Distribución de emisiones por tipo de fuente (1925–2021)",
              labels={"value": "Emisiones (MtCO₂)", "variable": "Fuente"},
              template="plotly_white")

fig.show()
```

**Interpretación: Distribución de emisiones por tipo de fuente (1925–2021)**

- Carbón (Coal) y Petróleo (Oil); dominan claramente las emisiones globales, marcando una diferencia notable respecto a las demás fuentes. El carbón se mantiene como la fuente más importante en casi todo el periodo analizado.

- Gas empieza a crecer significativamente a partir de la década de 1950, consolidándose como una fuente cada vez más relevante, aunque todavía menor en comparación con carbón y petróleo.

- Cement y Flaring tienen una contribución relativamente baja en el total. En particular: 
  - Cement presenta un incremento leve desde los años 80, ligado a la expansión de la construcción. 
  - Flaring mantiene una tendencia estable y con bajo impacto en todo el periodo.

En conjunto, este gráfico refleja que, pese a la diversificación energética, las emisiones históricas están concentradas principalmente en el uso de combustibles fósiles tradicionales.

**Distribución (%) de fuentes de emisiones de CO₂ (1925-2021)**

```python
# Columnas de fuentes
fuentes = ["Coal", "Oil", "Gas", "Cement", "Flaring"]

# Agrupar por año sumando emisiones globales
df_fuentes = df_1925.groupby("Year")[fuentes + ["Total"]].sum().reset_index()

# Calcular proporciones (%)
for fuente in fuentes:
    df_fuentes[fuente + "_pct"] = df_fuentes[fuente] / df_fuentes["Total"] * 100

# Gráfico de áreas apiladas
fig = px.area(
    df_fuentes,
    x="Year",
    y=[f + "_pct" for f in fuentes],
    title="📊 Distribución (%) de fuentes de emisiones de CO₂ a lo largo del tiempo",
    labels={"value": "Proporción (%)", "variable": "Fuente", "Year": "Año"}
)

fig.update_layout(yaxis_title="Porcentaje (%)", legend_title="Fuente de emisión")
fig.show()
```

**Observaciones relevantes**:

- **1925–1940:**
El carbón (Coal) domina claramente la composición de emisiones, representando en promedio cerca del 80 % del total mundial. Muy por detrás se ubica el petróleo (Oil), con una participación que oscila entre 25 % y 30 %, mientras que las demás fuentes son marginales.

- **Décadas de 1940–1960:**
A partir de los años 40, el porcentaje de emisiones provenientes del carbón comienza a disminuir gradualmente, mientras que las de petróleo (Oil) y, en menor medida, gas (Gas) aumentan su participación. Las fuentes Cement y Flaring se mantienen prácticamente constantes y con peso muy bajo.

- **Décadas de 1960–2000:**
Hacia mediados de los 60, la proporción del carbón cae por debajo del 50 %, impulsada por el fuerte crecimiento del petróleo, que se convierte en la principal fuente de emisiones de CO₂ a nivel global, superando el 40% a mediados de los 70. El gas natural también gana relevancia, aunque en una magnitud menor.

- **Desde 2000 en adelante:**
A comienzos del siglo XXI, el carbón retoma el liderazgo en las emisiones de CO₂, superando nuevamente al petróleo en participación global. En tercer lugar se ubica el gas, que continúa aumentando sus emisiones, mientras que cement y flaring se mantienen a gran distancia, con contribuciones pequeñas pero en crecimiento sostenido.

## ⚖️ 🇨🇳 vs 🇺🇸 Comparación China – EE.UU. (1925–2021)

**Evolución de CO₂: China vs USA (1925–2021)**

```python
#Evolución de emisiones de CO₂: China vs USA (1925–2021)

# Filtrar solo los dos países
df_comparacion = df_1925[df_1925["Country"].isin(["China", "USA"])]

# Agrupar por año y país, sumando emisiones totales
df_comparacion = df_comparacion.groupby(["Year", "Country"])["Total"].sum().reset_index()

# Gráfico de líneas
fig = px.line(df_comparacion,
              x="Year",
              y="Total",
              color="Country",
              title="📈 Emisiones de CO₂: China vs USA (1925–2021)",
              labels={"Total": "Emisiones (MtCO₂)", "Year": "Año"})

fig.show()
```

### China vs US Emisiones de CO₂ (1925–2021)

**1925–1950**

- Estados Unidos presenta emisiones muy superiores a China, con una línea ascendente.

- En este periodo China prácticamente no muestra crecimiento significativo.

- Se observa una leve caída de emisiones CO₂ durante los años 30 por parte de EE.UU., vinculada a la Gran Depresión.

**1950–2000**

- EE.UU. mantiene un claro predominio en emisiones.

- A mediados de los 50, China comienza un crecimiento progresivo de sus emisiones.

- Desde la década de 1970, la curva de China acelera su pendiente, marcando el inicio de un crecimiento más sostenido.

**2000 en adelante**

- China experimenta un crecimiento exponencial, mientras que EE.UU. entra en una fase de estabilización con ligera tendencia a la baja.

- El 2006 se produce el punto de convergencia: China supera por primera vez a EE.UU. en emisiones de CO₂.

- Desde entonces, China continúa aumentando (con excepciones puntuales como 2005), consolidándose como el mayor emisor global.

📌 **Conclusión**: 
- El gráfico refleja el **traspaso del liderazgo en emisiones**:  
**Estados Unidos** dominó el siglo XX como principal emisor, mientras que **China** asumió ese rol en el siglo XXI, impulsado por su acelerada industrialización y urbanización.  
- Este desplazamiento del liderazgo en emisiones no solo marca un cambio en la **geografía económica mundial**, sino también en los **desafíos de la gobernanza climática global**, donde el equilibrio entre desarrollo y sostenibilidad se vuelve cada vez más crucial.

**Comparación de emisiones por década (barras lado a lado)**

```python
# Definir países a comparar
paises = ["China", "USA"]

# Crear columna de década
df_1925["Decada"] = (df_1925["Year"] // 10) * 10
# filtrar y agrupar
df_decada = df_1925[df_1925["Country"].isin(paises)].copy()
df_decada["Década"] = (df_decada["Year"] // 10) * 10
df_decada = df_decada.groupby(["Década", "Country"])["Total"].sum().reset_index()
# Gráfico de barras
fig = px.bar(df_decada, x="Década", y="Total", color="Country", barmode="group",
             title="📊 Emisiones totales por década: China vs USA",
             labels={"Total": "Emisiones (MtCO₂)"})

fig.show()
```

**Evolución de emisiones por fuente (líneas múltiples)**

```python
# Filtrar países y fuentes
paises = ["China", "USA"]
fuentes = ["Coal", "Oil", "Gas", "Cement", "Flaring"]

df_fuentes = df_1925[df_1925["Country"].isin(paises)]

# Agrupar por país, año y fuente (forma larga)
df_long = df_fuentes.melt(id_vars=["Country", "Year"], value_vars=fuentes,
                          var_name="Fuente", value_name="Emisiones")

# Gráfico
fig = px.line(df_long, x="Year", y="Emisiones", color="Fuente", facet_col="Country",
              title="📈 Evolución de emisiones por fuente: China vs USA",
              labels={"Emisiones": "MtCO₂"})

fig.show()
```

🧩 Análisis del gráfico

- **China:**
A partir de 1950, las emisiones provenientes del carbón (Coal) comienzan a incrementarse de manera sostenida, con una pendiente claramente ascendente. Este crecimiento se vuelve casi exponencial desde el año 2000, coincidiendo con la expansión industrial y urbana del país.
En contraste, las demás fuentes (Oil, Gas, Cement, Flaring) presentan incrementos más moderados y graduales, siendo el petróleo (Oil) la segunda fuente más destacada desde la década de 1970, aunque sin superar los 2000 MtCO₂.

- **Estados Unidos:**
Las emisiones de carbón (Coal) muestran un comportamiento más irregular: fueron más elevadas que las de China hasta mediados de los años 80, pero desde 2008 presentan una clara tendencia a la baja, reflejando la sustitución del carbón por gas natural y energías más limpias.
El petróleo (Oil) mantiene niveles superiores a los de China durante todo el periodo, mientras que el gas (Gas) tiene un papel más destacado que en China, con valores que superan los 1000 MtCO₂ en la década de 1970, aunque sin sobrepasar los 2000 MtCO₂.
Las emisiones por Cement y Flaring se mantienen constantes y con escasa variación, reflejando su menor peso relativo en la matriz energética estadounidense.

  - Con la **Revolución Industrial** (1850–1900) comenzó un crecimiento sostenido, especialmente por el uso del carbón.
  - A partir de 1950, las emisiones globales exhiben un crecimiento casi exponencial, con breves descensos asociados a crisis económicas o eventos globales (por ejemplo, 1973 y 2020).

- **Principales emisores históricos** 
  - EE.UU. y China se consolidan como los grandes dominadores en términos de emisiones acumuladas de CO₂. EE.UU. lideró durante más de un siglo debido a su temprana industrialización, alto consumo energético y expansión económica.

  - China asumió un papel protagónico en las últimas décadas (siglo XXI) con un crecimiento acelerado impulsado por su industrialización acelerada y su fuerte dependencia del carbón como fuente energética.

  - Europa tuvo un rol pionero en las emisiones históricas: 
    - Alemania y Reino Unido fueron actores clave durante las primeras etapas de la industrialización. Hoy, sin embargo, representan una proporción mucho menor en el contexto global, en parte por transiciones energéticas y políticas de reducción de carbono.
    - Francia destaca con un perfil energético más diversificado, con menor proporción de CO₂ gracias a la energía nuclear.

  - Rusia ha mantenido una presencia relevante, especialmente en la era soviética y posterior (cómo republica), vinculada al uso intensivo de carbón, petróleo y gas.

  - Canadá, aunque no se ubica entre los máximos emisores, también figura de forma consistente, reflejando su papel como economía industrializada con alta dependencia de recursos fósiles. 

- **Comparación China vs USA (1925–2021)**  
  - Hasta mediados del siglo XX, EE.UU. fue el principal emisor global, con amplia ventaja sobre China.

  - Desde la década de 1970, China comenzó un crecimiento sostenido en sus emisiones, vinculado a su industrialización, urbanización y demanda energética.

  - En 2006, se produjo el punto de quiebre: China superó a EE.UU. en emisiones anuales, marcando un cambio estructural en la geografía de emisiones de carbón a nivel global.  

  - Mientras que las emisiones de EE.UU. tienden a estabilizarse e incluso descender levemente, China mantiene una trayectoria ascendente, reflejando también la transición del liderazgo económico y energético global hacia el continente asiatico.

- **Fuentes de emisiones**  
  - **Carbón (Coal)** ha sido históricamente la fuente más dominante, con un repunte acelerado a comienzos del año 2000.  
  - **Petróleo (Oil)** tuvo un crecimiento marcado desde mediados del siglo XX, aunque con algunas caídas en los 70 y 80.  
  - **Gas** muestra un crecimiento sostenido desde 1950, consolidándose como una fuente relevante en la transición energética.  
  - **Cement y Flaring** tienen un impacto comparativamente menor, aunque como fuente Cement, presenta un leve aumento desde la década de 1980.

### 📌 Síntesis
 - El peso de las emisiones globales pasó de un dominio inicial de EE.UU. en el siglo XX a un liderazgo actual de China en el XXI, lo que refleja también los cambios en la economía mundial.

 - Las fuentes fósiles tradicionales (carbón, petróleo y gas) siguen siendo protagonistas absolutas, evidenciando la persistente dependencia de estos recursos a pesar del avance de las energías renovables.

 - Los mayores incrementos en emisiones se asocian a periodos de industrialización, urbanización y expansión económica, mostrando la estrecha relación estructural entre desarrollo y emisiones.

 - En el escenario actual, se observa una estabilización relativa de las emisiones en economías desarrolladas (como EE.UU. y Europa) frente a un ascenso sostenido en países emergentes, especialmente China, lo que plantea desafíos globales en términos de equidad y responsabilidad compartida en la reducción de emisiones.

Estas conclusiones constituyen una base sólida para discutir **políticas energéticas, estrategias de mitigación y compromisos internacionales** frente al cambio climático, donde la cooperación entre países desarrollados y emergentes será clave para lograr una transición efectiva hacia una economía baja en carbono.
